```python
import os
from pydub import AudioSegment

srcdirectory = 'mp3'
dstdirectory = 'wav'

# os.mkdir(dstdirectory)

for filename in os.listdir(srcdirectory):
    src = os.path.join(srcdirectory, filename)
    
    basename = filename.split(".")[0]
    dst = os.path.join(dstdirectory, f"{basename}.wav")
    
    
    sound = AudioSegment.from_mp3(src)
    sound.export(dst, format="wav")
```

```python
# check the number of wavfiles
wav_files = os.listdir("wav")
i = 0
for wav_file in wav_files:
  i += 1
print("number of wav files ", i)
```
