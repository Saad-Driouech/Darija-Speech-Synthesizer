```python
import csv
import os
import shutil

first_name = "Loubna"
last_name = "El hafidi"
first_name_ns = first_name
last_name_ns = last_name

if len(first_name.split()) > 1:
    first_name_ns = first_name.replace(" ", "")
    
if len(last_name.split()) > 1:
    last_name_ns = last_name.replace(" ", "")
```

```python
directory = f"Audios{first_name_ns}{last_name_ns}"
os.mkdir(directory)
```

```python
csv_file = open("annotatedSentencesAll.csv", "r", encoding="utf-8")
reader = csv.reader(csv_file)

new_csv_file = open(f"annotatedSentences{first_name_ns}{last_name_ns}.csv", "a", encoding="utf-8", newline="")
writer = csv.writer(new_csv_file)

header = ["num","id","sentence","path","url","firstname","lastname"]
writer.writerow(header)

for row in reader:
    if row[5] == first_name and row[6] == last_name:
        writer.writerow(row)
        src = f"Audios/{row[3]}"
        dst = f"{directory}/{row[3]}"
        shutil.copyfile(src, dst)
        
csv_file.close()
new_csv_file.close()
```
